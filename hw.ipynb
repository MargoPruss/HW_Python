```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Условие 1: задача 1

### 1.1 Считать файл

```python
df = pd.read_csv('kc-house-data.csv')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 6))
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 9))
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

## Условие 2: задача 2

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()

labels = ['Нет вида', 'Есть вид']
plt.pie(data, autopct='%1.1f%%', labels=labels)
plt.title('Распределение домов от наличия вида на набережную');
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data)
plt.title('Распределение этажей')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во домов')
plt.xticks();
```

В основном дома с 1 и 2 этажами

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data)
plt.title('Состояние домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во домов');
```

Большинство домов оценены по состоянию в 3 балла

- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. \
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше площадь дома, тем выше его стоимость

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Чем выше качество конструкции и дизайна дома, тем выше его стоимость

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Состояние дома')
plt.title('Влияние оценки дома на стоимость');
```

Наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Кол-ва этажей')
plt.title('Влияние кол-ва этажей на стоимость');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Вид')
plt.title('Влияние вида на стоимость');
```
